**Convert Categorical to Numerical using map**

```python
#Find the columns that are categorical (object) type

visa_df.select_dtypes(include='object').columns
```

```python
#Select the column and get unique labels
visa_df['case_status'].unique()
```

```python
#Assign numerical values to the unique labels according to the alphabetical order and put them in dictionary
d={'Certified':0,'Denied':1}
```

```python
#map the dictionary to the selected categorical column
visa_df['case_status']=visa_df['case_status'].map(d)
```

```python
#To convert all the categorical data to numeric data:
visa_df=pd.read_csv(r'C:\Users\User\Downloads\Visadataset.csv')
cols=visa_df.select_dtypes(include='object').columns
for col in cols:
    labels=list(visa_df[col].unique())
    labels.sort()
    vals=[i for i in range(len(labels))]
    d=dict(zip(labels,vals))
    visa_df[col]=visa_df[col].map(d)

visa_df
```

**Converting categorical to numerical using Label encoder**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
visa_df=pd.read_csv(r'C:\Users\User\Downloads\Visadataset.csv')
visa_df['case_status']=le.fit_transform(visa_df['case_status'])
visa_df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

visa_df=pd.read_csv(r'C:\Users\User\Downloads\Visadataset.csv')
cols=visa_df.select_dtypes(include='object').columns
for col in cols:
    visa_df[col]=le.fit_transform(visa_df[col])

visa_df
```

```python
#inverse_transform - will give the original values for the given column
print(visa_df['case_status'].values) # this gives the column values in an array
le.inverse_transform(visa_df['case_status'])
```

**Converting categorical to numerical using np.where**

```python
import numpy as np
visa_df=pd.read_csv(r'C:\Users\User\Downloads\Visadataset.csv')
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```
