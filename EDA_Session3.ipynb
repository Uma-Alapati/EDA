$take$

- take is used to extract the particular index of data
- take has one parameter called axis
- axis=1 means columns
- axis =0 means rows, default value
  

```python
visa_df.take([1,2],axis=1)
```

**At a time we cannot take both rows and columns. we need to do take twice as below.**

```python
visa_df.take([100,200,300]).take([2,4,5],axis=1)
```

$#important$

$iloc$

- Above drawback of take can be achieved by iloc.


```python
#visa_df.iloc[<rows>,<columns>]
visa_df.iloc[[100,200,300],[2,5,7]]
```
