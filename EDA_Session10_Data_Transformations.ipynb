**Data transformations**
  
- Generally data has 3 types of distributions
    - Normal distribution
    - Postive Skewed
    - Negative Skewed
- All the maths concepts developed by assumption as Data follows Normal distribution
- Data skew happens because of many reasons, in that one reason is ouliers
- Once we perform the outlier analysis then we can apply transformation methods , if the data does not follows normality
- We have some transformation methods mentioned below
    - Log Transformation
    - Exponential Transformation
    - Reciprocal Transformation
    - Square root Transformation
    - Power Transformation

**Log transformation**

- Log transformation means performing logarithm 
- Log means natural logarithm base=e

```python
x=2
np.log(x)
```

```python
exp_data=np.random.exponential(size=10000)
exp_data
```

```python
plt.hist(exp_data,bins=30)
plt.show()
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=30,label='Logarithm')
plt.show()
```

**Reciprocal Transformations**

- Reciprocal will fail when values are low
- For example x=0, then 1/0 will fail

```python
x=2
1/x
```

```python
reciprocal_data=np.reciprocal(exp_data)
reciprocal_data
```

```python
plt.hist(reciprocal_data,bins=30,label='Reciprocal')
plt.show()
```

**Square root transformation**


```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square Root')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
```

**Power Transformation**


- Box  : only positive data

```python
from sklearn.preprocessing import PowerTransformer

# Initialize the PowerTransformer
pt = PowerTransformer()

# Fit and transform the data
transformed_data = pt.fit_transform(exp_data.reshape(-1,1))

plt.hist(transformed_data,bins=30,label='Power Transformation')
plt.legend()
plt.show()
```
